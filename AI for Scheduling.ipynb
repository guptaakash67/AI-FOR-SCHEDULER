## History of Recruiting 

![alt text](https://img.etimg.com/photo/60986101/pg-4-1.jpg "Logo Title Text 1")
![alt text](https://img.etimg.com/photo/60986105/pg-4-2.jpg "Logo Title Text 1")

-  HR teams and recruiting managers have come a long way since the days of paper, faxed, and emailed resumes. 
-   Pre 90s: paper resumes, manual data uploads, inefficient and expensive job fairs. 
-  1999: Monster.com and Indeed.com: these were the first aggregators on the internet. They helped people find jobs, but didn’t do a lot for the companies who were hiring. 
-  Early 2000s: People looked for jobs on company websites, and some applied through company web portals, but these were difficult to navigate. 
- 2008: The financial downturn changed things for recruiting, companies had to be more selective as they evaluate applicants. 
- 2010- Today: Lots of available data on candidates. AI can be used on it to provide hiring insights for the first part of the pipeline
- AI can be used to schedule suitable candidates for the second part of the pipeline. 

#### Major Factors Affecting the recruiting industry

- Baby boomers are beginning to retire and vacate jobs.
- SaaS is on the rise again, building a second tech boom. 
- Social Media is becoming commonplace
- Younger generations entering the workforce no longer use newspapers and classifieds to find jobs, but online job boards, social media, and company recruiting websites.

##### Job fairs and in-person resume drop-offs are a thing of the past, but the volume of resume uploads increases. Finding the perfect candidate or even a handful of interview-ready candidates among the hundreds or thousands of job applications sent every month can be overwhelming.

- Recruiting software provides intelligent search and filtering functions to help reduce the number of resumes that recruiters actually have to read for a position. 
- i.e Search for degree types, keywords, and levels of experience within your online resume database, and find better candidates faster.

## Architecture

###### Our AI bot will use the following technologies

![alt text](https://files.readme.io/a769bab-API-AI_key-concepts.png "Logo Title Text 1")

![alt text](http://www.3bluefrogs.com/wp-content/uploads/2016/01/1452909989test1-1024x453.jpg "Logo Title Text 1")

- API.AI - The ai/machine learning tool used to process natural language
- Slack - Chat application used for RTM and interactive messages
- Google Calendar API - Used by bot to schedule meetings and reminders
- Express - Backend Server

##### Steps

1. When the slack bot is sent a message like "Schedule a meeting with a person at time on date or today/tomorrow/etc to discuss a meeting topic for one hour"  it is immedaitely sent to API.AI where multiple intents are already set up to parse the input and decide how to respond based off of the users input. 
2.  The scheduler bot will then check to see if it has permissions to add events to both users calendar. If not, it'll ask
3.  It will then check both users calendars to see if there any prexisting conflicts with any of the users calendars. If there are, the bot will then respond, letting the user know there are time conflicts, and then will prompt the user to pick a new time from a list of slots. 
4. The user will select the best time and the bot will then schedule the meeting, adding the event to all users calendars and sending each user and email.

- It can also create reminders for users using the same process as scheduling, the bot will set a reminder in the users calendar.
- This is done through the script.js file, Heroku's script scheduler, acting as a cron job to fire every night at midnight.
